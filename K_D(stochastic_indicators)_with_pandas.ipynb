```python
# for fast and slow k Stochastic Oscillator
def calculate_k_d(df, period=14):
  highs = df['high']
  lows = df['low']
  closes = df['close']

  period_maxs = highs.rolling(window=period).max()        #find max of a particular rolling window
  period_mins = lows.rolling(window=period).min()         #find min of a particular rolling window

  k = 100*((closes-period_mins)/(period_maxs-period_mins))
  d = k.rolling(window=3).mean()


  return k,d
```

```python
k, d = calculate_k_d(df, period=14)                       #default 14, alterable
```

```python
#present these side by side in dataframe
indicators = pd.concat([k, d], axis=1)
```
