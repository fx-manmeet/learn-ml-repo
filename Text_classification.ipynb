####This is very imbalanced dataset so it is better to use the Complement Naive Bayes classifier.

As it was designed to correct the “severe assumptions” made by the standard Multinomial Naive Bayes classifier. **It is particularly suited for imbalanced data sets.**

```python
X = df['Message']
y = df['Category']
```

```python
len(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)
```

```python
len(X_train)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import ComplementNB, MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score
```

```python
pipeMNB = Pipeline([('tfidf', TfidfVectorizer(stop_words='english')), ('clf', MultinomialNB())])
```

```python
pipeCNB = Pipeline([('tfidf', TfidfVectorizer()), ('clf', ComplementNB())])
```

```python
pipeSVC = Pipeline([('tfidf', TfidfVectorizer()), ('clf', LinearSVC())])
```

```python
pipeMNB.fit(X_train, y_train)
predictMNB = pipeMNB.predict(X_test)
```

```python
MNB_accuracy = accuracy_score(y_test, predictMNB)
```

```python
pipeCNB.fit(X_train, y_train)
predictCNB = pipeCNB.predict(X_test)
CNB_accuracy = accuracy_score(y_test, predictCNB)
```

```python
pipeSVC.fit(X_train, y_train)
predictSVC = pipeSVC.predict(X_test)
SVC_accuracy = accuracy_score(y_test, predictSVC)
```

```python
print(f'MultinomialNB accuracy: {MNB_accuracy}')
print(f'ComplementNB accuracy: {CNB_accuracy}')
print(f'LinearSVC accuracy: {SVC_accuracy}')
```

```python
print(classification_report(y_test, predictMNB))
```
